### Import libriaries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("housing.csv")
```

### Check headers, data types and get a view of the data set.

```python
data.info()
```

```python
# Sort the data on 'median house values' and display the 10 first rows
data.sort_values('median_house_value', ascending=False).head(10)
```

```python
# display the last 10 rows
data.tail(10)
```

### Check for missing values and drop coresponding rows

```python
# Check for missing values
print(data.isnull().sum())

# Remove rows with empty columns
data.dropna(inplace=True)

# Check again
print(data.isnull().sum())
```

### Descriptive analysis

```python
data.describe()
```

### Check the distribution of the median house values

```python
# Creating the figure
plt.figure(figsize=(10, 6))

# Creating the graph and some cosmetics
plt.hist(data['median_house_value'], bins=50, color='green', alpha=0.7)

# Title and labels
plt.title('Distribution of Median House Values')
plt.xlabel('Median House Value')

# Figure text (analysis of the data)
plt.figtext(0.2, -0.03, 'Analysis: House values appear to be concentrated towards the lower end. There is however a large \nquantity at 500.000 which might indicate that this is the limit of records', horizontalalignment='left')
plt.show()

```

### Correlation between median income and house values

```python
# Creating the figure
plt.figure(figsize=(10, 6))

# Creating the scatter plot and assigning data and cosmetics
plt.scatter(data['median_income'], data['median_house_value'], alpha=0.5, color='green')

# Title and labels
plt.title('Median Income vs. Median House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')

# Figure text (analysis of the data)
plt.figtext(0.2, -0.03, 'Analysis: There is a positive correlation between median income and median house values \nwhere areas with higher income tend to have higher house values', horizontalalignment='left')
plt.show()
```

### Comparing Analysis
#### Geografical distribution of prices

```python
# Creating the figure
plt.figure(figsize=(12, 8))

# Creating a scatter plot using 'longitude' and 'latitute' data and color the plots based on house value acording to a color map.
scatterplot = plt.scatter(data['longitude'], data['latitude'], c=data['median_house_value'], cmap='seismic')

# Adding a color bar
plt.colorbar(scatterplot, label='Median house value')

# Title and labels
plt.title('Geografical distribution of prices')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Figure text (analysis of the data)
plt.figtext(0.3, 0.03, 'Analysis: Prices are generally higher along the costal area', horizontalalignment='left')
plt.show()
```

#### House value, age and income vs. Ocean proximity

```python
# Creating a figure containing two subplots with the dimension 1,3.
fig, ax = plt.subplots(1,3, figsize=(20, 5))

# Group data to new variable (median house value mean) by ocean proximity
median_house_value = data.groupby('ocean_proximity')['median_house_value'].mean()

# Creating the left graph
ax[0].bar(median_house_value.index, median_house_value.values)

# Title and labels
ax[0].set(title='Median House Value', xlabel='Ocean Proximity', ylabel='Median House Value')


# Group data to new variable (median income mean) by ocean proximity
mean_median_income = data.groupby('ocean_proximity')['median_income'].mean()

# Create the middle graph
ax[1].bar(mean_median_income.index, mean_median_income.values)

# Title and labels
ax[1].set(title='Median Income', xlabel='Ocean Proximity', ylabel='Median Income')

# Group data to new variable (median income mean) by ocean proximity
housing_median_age = data.groupby('ocean_proximity')['housing_median_age'].mean()

# Create the right graph
ax[2].bar(housing_median_age.index, housing_median_age.values)

# Title and labels
ax[2].set(title='House age', xlabel='Ocean Proximity', ylabel='House age')


# Figure text (analysis of the data)
plt.figtext(0.3, -0.09, 'Analysis: House values are higher along the costal area and highest on islands which also hold the oldest houses.\nIn contrast to house value and age, median income is lowest on islands. ', horizontalalignment='left')
plt.show()
```
